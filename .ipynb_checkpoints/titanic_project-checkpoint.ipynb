** Handling missing values**  
For the exploratory phase of the data analysis, we will use the groupby() method for data frames (creating groupby objects) to obtain any metrics, such as counts, for groups of interest (e.g., total number of female and male passengers). This has some consequences on handling with NaNs.
Groupby objects will drop NaN groups (the group itself is NaN). If grouped by two variables the group will be dropped if one group variable is NaN (e.g., (var1,NaN)). The count() or mean() method for groupby objects will return the count/mean for non-NaN entries. As such, it is not required to remove NaNs from the dataset before when using groupby objects in combination with either of the two methods. 

** Some specifications **  
In general, the "PassengerId" column of the titanic dataset will be used to return metrics (e.g. counts) from groupby object-groups of interest. The reason is that this column does not contain any NaN entries, so we will always obtain the maximum number of non-NaN entries for any group.

In general, if a group has all NaN entries, it will be omitted in plots created via matplotlib library (invoked by the **plot_group_survival function** or by using the plot() method for data frames. So again, NaNs will not cause errors when the cognate group is plotted and, thus, don't require prior removal from the dataset.

However, special handling with NaNs is required with the **survival_by_group function**, which returns group-specific survival rates (pandas series). When used to return survival rates for age groups, it is required that missing groups (all NaN entries), are removed explicitly. "Age groups" will be defined via the .cut() method for data frames, introducing a "category"-type pandas series as an "Age group" column in the titanic data frame. Creating a groupby object for a "category"-type pandas series and applying the count() method will, however, retain missing groups!!
This would conflict with survival rates for groups with no survivors (0%), because  both, missing groups and 0% survival rate groups, would be represented as NaN without the possibility of differentiation!! Thus, the groupby objects will be first filtered for non-NaN groups, before survival rates are calculated, ensuring that only groups without any survivors will be represented as NaN. These NaNs will converted to 0 (via fillna(0) method)

```python
'''
The first three functions are helper functions (used with .apply() method for pandas series) to rename 
entries in particular columns in the titanic_csv dataset. This will make plotting more convenient.

e.g. no manual labeling in plots necessary, because the renamed entries will be used as labels in plots. Automatic 
labeling will reduce errors.
'''
def harbour_classification(entry):
    if entry == "S":
        return "Southampton"
    if entry == "Q":
        return "Queenstown"
    if entry == "C":
        return "Cherbourg"
    else:
        return entry
    
def ticket_classification(entry):
    if entry == 1:
        return "class 1 ticket"
    if entry == 2:
        return "class 2 ticket"
    if entry == 3:
        return "class 3 ticket"
    else:
        return entry
    
def survival_classification(entry):
    if entry == 1:
        return "survived"
    if entry == 0:
        return "not survived"
    else:
        return entry
    

titanic_csv["Embarked"] = titanic_csv["Embarked"].apply(harbour_classification) 
titanic_csv["Pclass"] = titanic_csv["Pclass"].apply(ticket_classification) 
titanic_csv["Survived"] = titanic_csv["Survived"].apply(survival_classification) 


def counts_groups(group):
    
    '''
    Function will group the titanic_csv dataset by a specified column and return a pandas series with 
    passenger counts for each group and group names as indexes. indexes can be used as labels for plotting.
    
    group (string): name of the column from the titanic dataset to be used for grouping 
    '''
    
    grouped = titanic_csv.groupby(group).count()["PassengerId"]
    return grouped

counts_embarked = counts_groups("Embarked")
counts_gender = counts_groups("Sex")
counts_ticket = counts_groups("Pclass")
counts_survival = counts_groups("Survived")

def plot_pie(data, name):
    '''
    Function will take a pandas series with counts and plot a pie chart with indexes as labels for the slices. Each relative
    slice percentage as well as the absolute values (in parentheses) will be displayed.
    
    data: pandas series with numerical values
    name: plot title (string)
    '''
    my_autopct = lambda pct: "{0:.1f}% \n ({1:.0f})".format(pct,(pct/100)*sum(data.values))
    plt.pie(data,labels = data.index, autopct = my_autopct)
    plt.axis('equal')
    plt.title(name)
```

```python
plt.figure(1, figsize =(14,7))
plt.suptitle('Figure 1', size =16)
plt.subplot(2,3,1)
plot_pie(counts_embarked, "Start Harbour") 

plt.subplot(2,3,2)
plot_pie(counts_gender, "Gender") 

plt.subplot(2,3,3)
plot_pie(counts_ticket, "Ticket class") 

plt.subplot(2,3,4)
plot_pie(counts_survival, "Survival") 
```

## Does the survival rate vary in different groups? 
** What was the survival rate for passengers embarking at different start harbours (Southampton, Queenstown and Cherbourg)? **  
Interestingly, the survival rate for passengers from Cherbourg was ~1.6x higher than for passengers from Southampton and ~1.4x higher than for passengers from Queenstown (Figure 2)  
** What was the survival rate for the different ticket classes? **  
Similarly, there are big differences in survival rates for the different ticket classes. First class passengers survival rate was 1.3x and 2.6x higher than the survival rate for second class and third class passengers, respectively (Figure 2).  
** What were the gender-specific survival rates? **  
Similar trends can be observed for the gender specific survival rates, with female survival rate being 3.9x higher than the male survival rate (Figure 2).

```python
def survival_by_groups(data, group):
    '''
    Function takes the titanic dataset, a column name from the dataset and 
    returns a pandas series with the survival rates (%) for cognate groups within the specified column.
    
    data: titanic dataset (pandas data frame)
    group: name of columns from the titanic dataset for which the survival rates should be calculated (list)
    '''
    
    total_in_group = data.groupby(group, as_index = True).count().loc[lambda df: df.PassengerId != 0]
    group.append("Survived")
    
    group_by_survival = data.groupby(group,as_index = True).count().reset_index(level = "Survived")
    survived = group_by_survival.loc[lambda df: df.Survived == "survived","PassengerId"].loc[total_in_group.index]

    
    return ((survived/total_in_group["PassengerId"])*100).fillna(0)


```

```python
survival_by_start_harbour = survival_by_groups(titanic_csv, ["Embarked"])   
survival_by_ticket_class = survival_by_groups(titanic_csv, ["Pclass"])   
survival_by_sex = survival_by_groups(titanic_csv, ["Sex"])   

def plot_group_survival(data,name):
    ''' 
    Function takes a pandas series and plots a horizontal bar chart. Row indexes will be set as y-axis labels
    
    data: pandas series object with row indexes containing survival values
    name: plot title (string) 
    '''
    
    average_survival_rate = (round(counts_survival.loc["survived"]/float( sum(counts_survival) ), 2))*100
    y_axis = range(len(data))
    plt.barh(y_axis,data,tick_label = data.index)
    average_survival = plt.axvline(average_survival_rate, color = "r")
    for i in range(len(data)):
        plt.text(10,i, "{}%".format(round(data.loc[data.index[i]],1)), verticalalignment='center' )
    plt.title(name)
    plt.xlabel("Survival rate (%)")
    plt.xticks(np.arange(0,101,10))
    plt.legend([average_survival], ["average survival rate"]) 
 
```

```python
fig, ax = plt.subplots(1,3, figsize =(16,5))

fig.add_subplot(1,3,1)
plot_group_survival(survival_by_start_harbour,"Survival by start harbour")
ax[0].set_xticks([])
ax[0].set_yticks([])

fig.add_subplot(1,3,2)
plot_group_survival(survival_by_ticket_class,"Survival by ticket class")
ax[1].set_xticks([])
ax[1].set_yticks([])

fig.add_subplot(1,3,3)
plot_group_survival(survival_by_sex,"Survival by gender")
ax[2].set_xticks([])
ax[2].set_yticks([])

plt.tight_layout()
fig.subplots_adjust(top=0.8)
fig.suptitle('Figure 2', size =16)
plt.show()
```

## Statistical analysis of the group-specific survival rates
Now let's conduct statistical tests and evaluate the group-specific survival rates in a statistical context, i.e do the observed differences provide enough evidence for a significant correlation with the investigated groups? Because we have nominal (categorical) variables and want to test for correlation between variables, we will conduct a  ** $\chi^2$ (chi-square) test for independence **.

\begin{equation*}
\chi^2 = \sum{\frac{f_o - f_e}{f_e}} 
\quad\quad\quad \text{$f_o$ = observed frequency} 
\quad\quad\quad \text{$f_e$ = expected frequency}
\end{equation*}


** Hypothesis **  
H0 (Null Hypothesis): There is no correlation between survival rate and one of the other variables **1. start harbour, 2.ticket class, 3. Gender**. Both variables are independent  
H1(Alternative Hypothesis): There is a correlation between survival rate and one of the other variables. The variables are not independent.

** Parameters for the chi-square test **  
**Survival by start harbour**  
critical statistic value (chi-square critical) = 5.991(alpha level = 0.05, degrees of freedom = 2)

** Survival by ticket class **  
critical statistic value = 5.991(alpha level = 0.05, degrees of freedom = 2)

** Survival by gender **  
critical statistic value = 3.841(alpha level = 0.05, degrees of freedom = 1)




```python
def chi_square_independence_test(data, group):
    '''
    Function takes the titanic dataset, a column name from the dataset and returns the result of a 
    chi-square test for independence for a specified variable (column name) and the "Survived" variable from
    the dataset.
    
    data: titanic dataset (data frame)
    group: name of the column from the titanic dataset to be used as specified variable (string)
    '''
    # creates a contingency table of the observed frequencies of variables "Survived" and "Specified Column"
    contingency_table_observed = pd.crosstab(index = data["Survived"], columns = data[group], margins = True)
    contingency_table_observed.index = np.append(contingency_table_observed.index[:-1], "coltotal")
    contingency_table_observed.columns = np.append(contingency_table_observed.columns[:-1], "rowtotal")
    
    rowtotal_not_survived = contingency_table_observed["rowtotal"].loc["not survived"]
    rowtotal_survived = contingency_table_observed["rowtotal"].loc["survived"]
    total = contingency_table_observed["rowtotal"].loc["coltotal"]
    
    # creates a contingency table of the expected frequencies of variables "Survived" and "Specified Column"
    # expected values are based on the marginal frequencies for the total row and column entries
    expected_not_survived = (contingency_table_observed.loc["coltotal"]*rowtotal_not_survived)/total
    expected_survived = (contingency_table_observed.loc["coltotal"]*rowtotal_survived)/total
    contingency_table_expected = pd.DataFrame()
    contingency_table_expected = contingency_table_expected.append([expected_not_survived,expected_survived,contingency_table_observed.loc["coltotal"]], ignore_index = True)
    contingency_table_expected.index = contingency_table_observed.index
    
    # calculates the chi-square statistic
    chi_square = sum( (((contingency_table_observed-contingency_table_expected)**2)/contingency_table_expected).loc["survived"] ) + \
    sum( (((contingency_table_observed-contingency_table_expected)**2)/contingency_table_expected).loc["not survived"] )
    
    # calculate Cramer's V as measure of association between the tested variables
    min_dimension = min(len(contingency_table_observed.columns[:-1])-1,len(contingency_table_observed.index[:-1])-1)    
    cramers_v = np.sqrt(chi_square/(total*min_dimension))
    
    print ("chi-square statistic for correlation between survival rate and {0}: {1}".format(group,round(chi_square,2))+\
           "\nCramer's V: {}".format( round(cramers_v,2) ))
```

** Results report on chi-square tests **

```python
chi_square_independence_test(titanic_csv, "Embarked")  
```

```python
chi_square_independence_test(titanic_csv, "Pclass")  
```

```python
chi_square_independence_test(titanic_csv, "Sex")  
```

The relation between the survival rate and each of the three variables (start harbour, ticket class, gender) is significant. Based on the chi-square test results, we reject the null hypothesis in favour of the alternative hypothesis.

relation between survival rate and start harbor, chi-square(2) = 26.49, p < .05, Cramer's V = 0.17  
relation between survival rate and ticket class, chi-square(2) = 102.89, p < .05, Cramer's V = 0.34   
relation between survival rate and gender, chi-square(1) = 263.05, p < .05, Cramer's V = 0.54   

This suggests that the probability of surviving was higher for female, first class ticket passengers and passengers from Cherbourg. Based on the guidelines for Cramer's V metric (Intro to inferential statistics, Udacity), the association between gender and survival rate exhibits a strong effect size, whereas the effect size for the association between survival rate and ticket class and survival rate and starting harbour is medium and small, respectively.


## What is the logic behind the different survival rates for the different groups?
** male versus female **  
The higher survival rate for women is in agreement with the "Women and children first" honorary code, by which the lives of women and children are supposed to be saved first. In this context let's check if a higher survival rate is also true for children. 

```python
def age_classification(bins, data):
    ''' 
    Function takes a sequence of numbers with a specified step size defining the age groups and returns
    a titanic dataset extended by an age group column that contains age group entries for each pasenger with known
    age. If a passenger age is missing, it will be classified as NaN age group.
        
        
    data:   titanic dataset (pandas data frame)
    
    bins:   sequence of numbers that define the range of age groups (list or tuple).
            for example: [0,5,10,15] would group passengers in classes of age 0-5, 5-10 and 10-15, respectively
    '''
    
    bins_grouped = pd.cut(data["Age"],bins)
    
    def filter_numerical(entry):
        filtering = filter(lambda ch: ch not in "(),[]", entry)
        return filtering
    bins_grouped = bins_grouped.apply(filter_numerical)
    
    def create_labels(entry):
        entry = entry.split(" ")
        return "age {0}-{1}".format(entry[0],entry[1])
    age_groups = pd.Series(bins_grouped.apply(create_labels), name = "Age group")
    
    return pd.concat([age_groups, data], axis = 1)
```

```python
bins = range(0,85,10)
titanic_with_age_groups_csv = age_classification(bins, titanic_csv)
survival_by_age = survival_by_groups(titanic_with_age_groups_csv, ["Age group"]) 


fig, ax = plt.subplots(1,1, figsize =(10,6))
fig.add_subplot(1,1,1)
plot_group_survival(survival_by_age,"Survival by age group")
ax.set_xticks([])
ax.set_yticks([])
fig.subplots_adjust(top=0.8)
plt.suptitle('Figure 3', size =16)
plt.show()
```

Indeed, passengers in the group aged from 0-10 years exhibit the highest survival rate (59.4%) amongst all age groups with an average 1.8 x higher survival rate than passengers from 10 years of age onward (Figure 3; survival rates for different age groups can be visualized by the user). 

Again, let's test for statistical significance for the correlation between age and survival rate by conducting a chi-square test for independence.

** Hypothesis **  
H0 (Null Hypothesis): There is no correlation between survival rate and age. Both variables are independent  
H1(Alternative Hypothesis): There is a correlation between survival rate and age. The variables are not independent.

**parameters for the chi-square test**  
chi-square critical = 14.067 (alpha level = 0.05, degrees of freedom = 7)

```python
chi_square_independence_test(titanic_with_age_groups_csv, "Age group")  
```

** Results report on chi-square test **  
The relation between survival rate and age is significant, chi-square(7) = 15.3, p < .05, Cramer's V = 0.15 (medium effect size). Children (Age group 0-10) were more likely to survive than passengers of age 10 years onward.

** ticket classes **  
In contrast, the differences in survival rates for the different ticket classes might be explained by a passenger's status, i.e. the crew prioritized first class passengers in terms of boarding rescue boats. Alternatively, first class cabins were simply closer to most rescue boats and thus, first class passengers were more likely to board rescue boards first. The ship had multiple decks (from top to bottom, A-G plus three additional; http://www.titanicfacts.net/the-titanic.html), of which obviously some were closer to the top deck, thus, closer to rescue boats. I assume that the character in each "Cabin" column entry (e.g. C85) in the titanic dataset represents the deck type. In this context, let's see if first and second class passengers resided predominantly on upper decks (A-C), whereas third class passengers resided on lower decks (D-).



```python
def deck_classification(entry):
    ''' 
    helper function to convert character-numerical entries of the "Cabin" column in the 
    titanic_csv into one letter code. The letter code coresponds to deck type.
    
    A -> deck A (topmost deck with cabins)
    .
    .
    .
    G -> deck G (lowermost deck with cabins)
        
    (e.g., C85 will be converted to C)
    '''
    if pd.isnull(entry):
        return np.nan
    else:
        if "A" in entry:
            return "A deck"
        if "B" in entry:
            return "B deck"
        if "C" in entry:
            return "C deck"
        if "D" in entry:
            return "D deck"
        if "E" in entry:
            return "E deck"
        if "F" in entry:
            return "F deck"
        if "G" in entry:
            return "G deck"
        
    
 

titanic_csv["Cabin"] = titanic_csv["Cabin"].apply(deck_classification)


def deck_types_by_ticket_class(data):
    ''' 
    This function will take the titanic dataset with the one letter code for the "Cabin" column, group it 
    by ticket class and deck type ("Pclass" and "Cabin) and calculate the passenger counts for each group. 
    It will return a data frame with ticket class, cabin types and counts as columns, and rows representing 
    passenger counts for each ticket class and cabin type.
    
    
    (e.g.: 1, A, 15 corresponding to 15 passengers with ticket class 1 on deck type A
        
    data: titanic_csv with cabin entries converted to one letter code (pandas data frame)
    '''
   
    
    grouped = data[["Pclass", "Cabin", "PassengerId"]].groupby(["Pclass", "Cabin"], as_index=False).count()
    
    # the set of deck types (in one letter code) is determined for the titanic dataset, against which the individual 
    # deck type sets of each ticket class are compared. If a deck type is missing from a ticket class, a value of 0 
    # is set for this deck type and a new entry is appended to the grouped titanic_csv (e.g.: 1, A, 0 corresponding
    # to 0 passengers with ticket class 1 on deck type A). 
    
    class1 = set( grouped[grouped["Pclass"] == "class 1 ticket"]["Cabin"] )  
    class2 = set( grouped[grouped["Pclass"] == "class 2 ticket"]["Cabin"] ) 
    class3 = set( grouped[grouped["Pclass"] == "class 3 ticket"]["Cabin"] )     

    dict_cabins = {"class1":class1, "class2":class2, "class3":class3}
    all_categories = class1 | class2 | class3

    for ticket_class in dict_cabins:
            for category in all_categories:
                if category not in dict_cabins[ticket_class]:
                    missing = pd.DataFrame([[ticket_class, category,0]], columns=grouped.columns)
                    grouped = grouped.append(missing)
                    
    def transform(entry):
        if entry == "class1":
            return "class 1 ticket"
        if entry == "class2":
            return "class 2 ticket"
        if entry == "class3":
            return "class 3 ticket"
        else:
            return entry
        
    grouped["Pclass"] = grouped["Pclass"].apply(transform)    
    return grouped


    
def transpose(data,index_col,transpose_col):
    '''
    takes a data frame of the titanic dataset that was grouped by two variables and contains numerical values 
    for each group. Each group and the corresponding numerical value must be represented as a one row entry. So when
    using the groupby() function for data frames, the argument as_index must be set to False. The "PassengerId" column
    from the titanic dataset must be used to represent the numerical values. 
    It returns a transposed data frame with the entries set of one of the two variables used for grouping as columns
    and the entries set of the other variable used for grouping set as row index. 
    
    As such, the returned data frame can be directly used to plot the number of passengers as a function of the 
    two columns used for grouping (via .plot() method for data frames).
    
    e.g., data frame with columns "Pclass" (ticket class), "Cabin" (cabin type), "PassengerId" (counts) will be 
    converted into data frame with columns "A", "B" and index "class1", "class2"
    
    Pclass    |    Cabin   |   PassengerId           ------->                    A   |   B
    ----------------------------------------                          ---------------------
    class 1   |     A      |       15                                 class 1    15  |   25
    class 1   |     B      |       25                                 class 2    0   |   10
    class 2   |     A      |       0
    class 2   |     B      |       10
    
    
    index_col: name of column from the titanic dataset to be used to set row indexes (string)
    transpose_col: name of column from the titanic dataset to be used for setting columns in the transposed df (string)
    '''    
    
    data = data.sort_values(by = [index_col, transpose_col])
    index = oset(data[index_col].values)
    
    
    # 1. create dictionary with the set of transpose_col entries as keys (columns for the new data frame) and 
    # the corresponding "PasserngerId" entries as values. Dictionary will be used to create transposed data frame.
    # 2. monitor length of each dict numpy array and keep track of max length for any column
    # 3. loop over columns dict and compare the length of each list with the max length
    # 4. if any column has less entries than column with most entries (len(column_i) < len(column_max)) append 0
    
    # This will make sure that transposed data frame will contain all groups with missing group values set as 0
    # (group by Gender and Cabin has a group  "female, G deck" but is missing the group "male, G deck". As such, an
    # entry = 0 will be set for the column male/ row index male). This is crucial as we need same column lengths 
    # to create a data frame!!
    
    columns = dict()
    all_categories = set(data[transpose_col])
    len_column = None
    max_len = None
    
    for colname in all_categories:
        columns[colname] = data.loc[lambda df: df[transpose_col] == colname,:]["PassengerId"].values
        len_column = len(columns[colname])
        if max_len < len_column:
            max_len = len(columns[colname])

    for key in columns:
        if len(columns[key]) != max_len:
            i = max_len-len(columns[key])
            for repeat in range(i): 
                columns[key] = np.append(columns[key],0)
                
        
    return pd.DataFrame(columns, index = index)    


counts_deck_ticket = deck_types_by_ticket_class(titanic_csv)

transpose(counts_deck_ticket,"Pclass","Cabin").plot(kind = "bar",rot = 1, figsize = (12,6), title = "Accomodation by ticket class and deck type")  
plt.subplots_adjust(top=0.8)
plt.ylabel("Number of passengers")
plt.suptitle('Figure 4', size =16)
plt.show()
```

Of the passengers who boarded with first class tickets 121 resided on decks A-C, whereas 54 resided on decks D and E (Figure 4).
In comparison to the total count of first class tickets (216; see figure 1), the data for first class deck occupancy is representative. So, indeed, it seems that the majority of first class passengers resided on upper decks, which might have provided a strategic advantage in reaching the rescue boats! Second class and third class passengers exhibit predominantly occupancy of decks D-G (lower decks), however, the cognate data is not at all representative, as only 16 out of 184 and 12 out of 491 data points (compare figure 1 and figure 4) are available for second and third class deck occupancy, respectively. 

But according to following sources (http://www.dummies.com/education/history/titanic-facts-the-layout-of-the-ship/, https://www.encyclopedia-titanica.org/cabins.html) most second and third class cabins were located on decks D,E,F (some third class cabins on deck G). So both classes seemed to have a disadvantage in terms of good accessibility of the boat deck harbouring rescue boats. But what is then a possible explanation for the discrepancy in survival rates between class 2 and class 3 (see figure 2)? We already provided a possible explanation for higher survival rates for women as opposed to men. Let's see if there is an imbalance in the female/ male proportion between class 2 and class 3.

```python
counts_gender_ticket = titanic_csv.groupby(["Pclass", "Sex"], as_index = False).count()

transpose(counts_gender_ticket,"Sex","Pclass").plot(kind = "bar", rot = 1, figsize = (12,6), title = "Passenger by Gender and ticket class")
plt.ylabel("Number of passengers")
plt.subplots_adjust(top=0.8)
plt.suptitle('Figure 5', size =16)
plt.show()
```

Indeed, we see that of the total third class tickets relatively more were purchased by men (figure 5). This is contrasted by a much more balanced female versus male ration for first and second class, respectively. Thus, the higher number of male passengers in third class might account for a overall lower survival rate for third class in comparison to second class (and first class).

In this context, the comparison of male vs. female survival rates for all classes could provides further insights if survival rate differences for ticket classes are solely correlated with differences in gender-specific occupancy rates , or if other variables might have also played an important role in determining a passenger's survival probability. Since we hypothesized an overall higher survival probability for women based on the "Women and children first" honorary code, one would expect similar survival rates for women across all three ticket classes if gender is the only explanatory variable.

```python
survival_by_gender_ticket = survival_by_groups(titanic_csv, ["Sex", "Pclass"]).reset_index()


transpose(survival_by_gender_ticket,"Sex","Pclass").plot(kind = "bar", rot = 1, figsize = (10,5), title = "Survival by gender and ticket class")
plt.subplots_adjust(top=0.8)
plt.ylabel("Survival rate (%)")
plt.suptitle('Figure 6', size =16)
plt.show()
```

As evident from figure 6, survival rates for female passengers in class 1 and class 2 were ~2x higher than the survival rate for females in class 3. So it is very likely that in addition to gender other variables influenced survival probabilities. One explanation might be cabin location (which decks).

Intriguingly, male first class passengers exhibit 2.4x higher survival rates than male passengers in class 2 (and class 3). Since the differences in female survival rates between class 1 and class 2 are only marginal (1.05 x higher for class 1), there must be yet an another explanatory variable in addition to cabin location. One possibility could be the passengers' status, as stated before.

** starting harbour **  
There is no logical explanation as to why a passenger's origin would have an influence on the probability of survival. Let's see if either of the two variables, gender and ticket class, which we already identified as being explanatory, can provide any information in the context of survival rates for the three different starting harbours (Cherbourg, Queenstown, Southampton; see figure 2). 

```python
counts_embarked_ticket = titanic_csv.groupby(["Embarked", "Pclass"], as_index = False).count()
counts_embarked_gender = titanic_csv.groupby(["Embarked", "Sex"], as_index = False).count()

fig, axes = plt.subplots(1, 2, figsize = (15,6))
axes[0].set_ylabel("Number of passengers")
axes[1].set_ylabel("Number of passengers")

fig.subplots_adjust(top=0.8)
transpose(counts_embarked_ticket,"Embarked","Pclass").plot(kind = "bar",rot = 1, ax = axes[0], title = "Passenger by Start Harbour and ticket class")
transpose(counts_embarked_gender,"Embarked","Sex").plot(kind = "bar", rot = 1, ax = axes[1], title = "Passenger by Start Harbour and Gender")
plt.suptitle('Figure 7', size =16)

plt.show()
```

For Cherbourg, which exhibits the highest survival rate (figure 2), we see that the numbers for first class and female passengers, respectively, are proportionally higher than the corresponding values for Southampton (figure 7), which shows the lowest survival rate (figure 2). Both, the first class tickets and females, were identified as being correlated with higher survival rates.
Queenstown, which exhibits a survival rate in between Cherbourg and Southampton, shows a ratio of female/male passengers similar to Cherbourg, but has a markedly lower fraction of first class passengers. The proportionally higher numbers of third class passengers might explain the slightly lower survival rate in comparison to Cherbourg (cabin location, as argued in the section above).

## Summary
Overall, the data shows that a passenger's survival probability is highly correlated to gender, age and ticket category, with gender showing the strongest association. Female passengers were almost 4x as likely as men to survive. One possible explanation is the higher priority in saving women, which is supported by the finding that children (age 0-10) as well were almost twice as likely to survive. This is known as the "Women and children first" honorary code in situations of abandoning ships (https://en.wikipedia.org/wiki/Women_and_children_first).
However, higher survival probabilities also correlate with more expensive ticket classes, which might be explained by cabin locations being on upper decks of the ship. In this context I hypothesize that rescue boots are easier and faster accessible for cabins located on the upper half of the ship. 


Nevertheless, there are also some limitations inherent to the titanic dataset. 
For example, entries in "Parch" and "SibSp" columns are highly ambiguous. The former describes if parents or children travelled together with the specified passenger, the latter defining related siblings or spouse (www.kaggle.com). The entries of each column simply state the number of related family members without providing any direct means for discriminating the different "classes" (type of family relation). For example, the first entry in the titanic_csv dataset, a passenger with the name "Braund, Mr. Owen Harris" has a 1 in the "SibSp" column. The only possibility to figure out if the entry is either related to sibling or spouse, would be to filter for the other passenger with the name "Braund". If the second "Braund" named passenger is male (Mr. in the "Name" column entry), the "SibSp" entry most likely referes to sibling (here, we assume that homosexual marriages were not allowed by the law at that time). In case of a female passenger (Miss. or Mrs. in the "Name" column entry), the abbreviation Miss. and Mrs. would point to sibling and wife, respectively. 
So it is required to view the family relation in the context of all family members (same name) in the dataset, in order to being able to correctly define each member (parents, children, spouse, sibling), which makes it very laborious to group by family relation and analyse the data in this context.

Missing values can be found for 3 variables, namely "Age", "Cabin" and "Embarked. Here, "Cabin" exhibits a significant loss of data, contributing only 22% of total cases (688 missing values for "Cabin"). Furthermore, the "Cabin" part of the dataset is highly skewed with almost 90% of cases representing data for first class ticket passengers (figure 4), which, as already stated in the results section, raises issues on inferred correlations with the survival rate. The titanic dataset does not provide enough information to define robust correlations between cabin location and survival rate. So it was required to obtain additional data (basic information on ship construction with deck-specific cabin locations; see results section for ticket classes) to support any conclusions on survival rates based on cabin location. One has to be aware of the interpretation being potentially biased and speculative, because the additional data is not directly related to the titanic dataset!!

Similarly, one has to be careful with the interpretation of the correlation between age group and survival rate. The reason is that numbers of cases strongly vary between groups, depending on the defined age group range (min and max age for each group). For example age group 75-80 exhibits 100% survival rate, however, including only 1 case!! The non-representative character of such data strongly reduces the reliability of any inferential analysis.  

** potential questions to address with more data available**  
In general, it would be very interesting to use similar data to evaluate potential measures in order to improve safety on ships and, thus, increase the overall survival probability in case of an accident at see. 
For example, having information on different causes of deaths, i.e. passengers who died directly "on the ship" (remained on the sinking ship) versus passengers who died of indirect causes such as injuries and severe hypothermia. In this very context, it would make sense to compare the corresponding times it took rescue teams to arrive at the scene of accident (and provide medical assistance for those in need) for different events. Is there a significant increase in survival rate with decreasing arrival time of rescue teams?  Furthermore, what is the metric of the same correlation at different air and water temperatures, i.e is fast rescue service especially important at unfavourable weather conditions?
Another relevant point in this context would be, what is the proportion of non-survivors who simply drowned because of not wearing a life jacket? This encompasses all passengers who were not trapped inside but actually managed to disembark a sinking ship.  
Similarly, it would be interesting to have more quantitative data on the location of cabins in relation to accessibility of emergency exits. What is the survival rate for different cabin distances to emergency exits? Is there a significant increase in overall survival rate for a certain average cabin distance to emergency exits (better accessibility of the boat deck)? Is there a positive correlation between corridor width and survival rate (wider corridors might decrease evacuation times)?

Finally, I assume that evacuation time might be a significant factor influencing overall survival probability with survival rate and evacuation time being inversely correlated (decreasing evacuation time relates to increasing survival rate). Of course, evacuation time depends on many factors, many of which are not easily "adaptable", such as number of passengers, number of old and very young passengers (I assume slower than average evacuation times for these groups) or awareness and response time of the crew. However, I am curious if the two different evacuation strategies, 1. few but high capacity recue vessels, 2. many but low capacity vessels, might have different impacts on the time it takes to evacuate a ship. 

Overall, all this data could be used to infer the most significant factors, and based on the analysis, one could elaborate a safety concept intended to maximize the overall survival probability in case of ship accidents.
